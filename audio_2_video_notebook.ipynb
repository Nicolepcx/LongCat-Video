```python
from pathlib import Path
import shutil, subprocess

import imageio_ffmpeg

img_src = Path("assets/avatar/custom/tabby.png")
audio_src = Path("assets/avatar/custom/audio.wav")

assert img_src.exists(), f"Missing: {img_src.resolve()}"
assert audio_src.exists(), f"Missing: {audio_src.resolve()}"

work = Path("assets/avatar/custom")
work.mkdir(parents=True, exist_ok=True)

# copy image to a fixed name
ref_img = work / "ref.jpg"
shutil.copy(img_src, ref_img)

# convert audio to 16 kHz mono wav (even if it already is wav)
ffmpeg = imageio_ffmpeg.get_ffmpeg_exe()
out_wav = work / "audio.wav"

cmd = [ffmpeg, "-y", "-i", str(audio_src), "-ar", "16000", "-ac", "1", str(out_wav)]
print("Running:", " ".join(cmd))
subprocess.check_call(cmd)

print("Staged image:", ref_img)
print("Staged audio:", out_wav)
```

```python
import json, pathlib

src = pathlib.Path("assets/avatar/single_example_1.json")
dst = pathlib.Path("assets/avatar/custom/at2v.json")
dst.write_text(src.read_text())

data = json.loads(dst.read_text())

def patch(obj):
    if isinstance(obj, dict):
        for k, v in obj.items():
            if isinstance(v, str):
                low = v.lower()
                if "audio" in low and any(ext in low for ext in [".wav",".mp3",".flac",".m4a",".aac",".ogg"]):
                    obj[k] = "assets/avatar/custom/audio.wav"
            else:
                patch(v)
    elif isinstance(obj, list):
        for it in obj:
            patch(it)

patch(data)

for key in ["prompt", "text", "caption"]:
    if key in data and isinstance(data[key], str):
        data[key] = "A realistic person speaking naturally, subtle head motion, stable identity, high quality"

dst.write_text(json.dumps(data, indent=2))
print("Wrote:", dst)
print(dst.read_text()[:600])
```
